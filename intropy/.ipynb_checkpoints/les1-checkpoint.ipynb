```python
#Calculate Mean,Median,Mode,STD 
import pandas as pd 
import matplotlib.pyplot as plt

coviddata = pd.read_csv('subset-covid-data.csv')
cases = coviddata.get('cases')
deaths  = coviddata.get('deaths')
print('Cases: Mean=',cases.mean(),'Median=',cases.median(),'STD=',cases.std())
print('Deaths: Mean=',deaths.mean(),'Median=',deaths.median(),'STD=',deaths.std())

max_case_index = cases.idxmax()
print('Country with max cases is ', coviddata.loc[max_case_index,'country'])
plt.figure(figsize=(15,8))
plt.hist(cases,bins=100)
plt.show()
```

```python
#Calculate Mean,Median,Mode,STD 
import pandas as pd 
data = pd.read_csv('Calories.csv')
Calories = data['Calories']
print('cal: Mean=',Calories.mean(),'Median=',Calories.median(),'Mode',Calories.mode()[0],'STD=',Calories.std())
```

```python
min_duration_row = data.loc[data['Duration'].idxmin()]
print("\nRow with minimum Duration:")
print(min_duration_row)
```

```python
#Relace empty row of calories with mean
data['Calories'].fillna(Calories.mean(),inplace=True)
print(data['Calories'])
```
